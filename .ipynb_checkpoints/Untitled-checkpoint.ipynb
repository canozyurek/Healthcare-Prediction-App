```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
data = data[['age', 'avg_glucose_level', 'bmi']]
```

```python
data.dropna(inplace=True)
```

```python
ss = StandardScaler()
```

```python
data_transformed = ss.fit_transform(data)
```

```python
inertia = []

for i in range(1, 10):
    clustering = KMeans(n_clusters=i, init='k-means++', random_state=42)
    clustering.fit(data_transformed)
    inertia.append(clustering.inertia_)
    
ks = np.arange(1, 10)

fig, ax = plt.subplots(figsize=(10, 8))
sns.lineplot(x=ks, y=inertia, ax=ax)
ax.set_xticks(ks)
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data_transformed)
```

```python
data_transformed = pd.DataFrame(data_transformed, columns=data.columns)
```

```python
data_transformed['clusters'] = kmeans.predict(data_transformed)
```
